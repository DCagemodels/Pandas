Seaborn with heatmaps

```python
import pandas as pd
import matplotlib.pyplot as plt 
%matplotlib inline
```

```python
import seaborn as sns
```

```python
df = pd.read_csv(r'C:/Users\draco\Desktop/Python Boot Camp/Jupyter/New_olympics.csv')
df.head()
```

Using the Olympic dataset, present a summary of the total medals won by participating countries in the 2008 Olympics

```python
lo = df[df.Edition == 2008]
lo
```

```python
g = lo.groupby(['NOC', 'Medal']).size().unstack('Medal',fill_value=0)
gg = g.sort_values(['Gold','Silver','Bronze'], ascending=False)[['Gold','Silver','Bronze']]
gg
```

```python
sns.heatmap(gg)
```

Transpose of data frame

```python
gg = gg.transpose()
gg
```

```python
plt.figure(figsize=(16,5))
sns.heatmap(gg)
```
